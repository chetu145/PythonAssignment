```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Group6_Final_Dataset.csv")

print("Initial Shape:", df.shape)
```

```python
# REMOVE DUPLICATES
df = df.drop_duplicates(subset="listing_id")
```

```python
# Replace missing numeric values with median
df["beds"] = df["beds"].fillna(df["beds"].median())
df["baths"] = df["baths"].fillna(df["baths"].median())

# Replace missing categorical values
df["property_type"] = df["property_type"].fillna("Unknown")
df["area"] = df["area"].fillna("Unknown")
df["eircode"] = df["eircode"].fillna("Unknown")
```

```python
# Remove unrealistic prices
df = df[(df["price"] > 50000) & (df["price"] < 10000000)]

# Remove unrealistic beds/baths
df = df[(df["beds"] > 0) & (df["beds"] < 10)]
df = df[(df["baths"] >= 0) & (df["baths"] < 10)]
```

```python
# Fix Data Type
df["price"] = df["price"].astype(int)
df["beds"] = df["beds"].astype(int)
df["baths"] = df["baths"].astype(int)
df["distance_from_city_km"] = df["distance_from_city_km"].astype(int)
```

```python
# Clean Text data
df["area"] = df["area"].str.strip()
df["property_type"] = df["property_type"].str.strip()
```

```python
# Create price category (for classification)

def price_category(price):
    if price < 400000:
        return "Affordable"
    elif price < 800000:
        return "Mid-Range"
    else:
        return "Expensive"

df["price_category"] = df["price"].apply(price_category)
```

```python
# Verify Data
print("\nCleaned Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nPrice Category Distribution:\n", df["price_category"].value_counts())
```

```python
# Save Cleaned Data
df.to_csv("Group6_Cleaned_Dataset.csv", index=False)

print("\n Dataset Saved")
```
